```python

tmp_date = tmp2['Dates']
tmp_date
```

```python
year=[]
for i in range(len(tmp_date)):
    year.append(tmp_date[i].split('-')[0])
```

```python
month = []
for i in range(len(tmp_date)):
    month.append(tmp_date[i].split('-')[1])
```

```python
len(year), len(month)
```

```python
df_year = pd.DataFrame(year)
df_month = pd.DataFrame(month)
```

```python
df_year.columns
```

```python
df_year=df_year.rename(columns={0:'year'})
df_month=df_month.rename(columns={0:'month'})
```
