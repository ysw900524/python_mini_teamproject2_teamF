```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import folium 
# from datetime import datetime
from pandas import Series, DataFrame
import datetime
```

```python
df = pd.read_csv('./data/Sanfrancisco/train.csv',encoding = 'UTF-8')
df["Dates"].head()
```

```python
results = []
for i in range(len(df)):
    a = df["Dates"][i]
    result = datetime.datetime.strptime(a,'%Y-%m-%d %H:%M:%S')
    result = result.strftime('%Y-%m')
    results.append(result)
```

```python
results[0]
```

```python
J = DataFrame(results)
df3 = pd.concat([J, df], axis = 1)
df3=df3.drop(["Dates","Descript",'DayOfWeek','PdDistrict','Resolution','Address'], axis= 1)
df3.columns = ["Date",'Category', 'X', 'Y']
df4 = df3[df3["Date"] == '2015-05']
df4
```

```python
df4['Category'].value_counts()
```

```python
map_3 = folium.Map(location=[37.77, -122.42], zoom_start=12, tiles='Stamen Terrain')

idx = 0.
for i in range(len(df4)):
    if df4["Category"][i] == 'LARCENY/THEFT':
        lati_val = abs(float(df4['Y'][i]))
        long_val = abs(float(df4['X'][i]))
        
#         lati_val = int(lati_val*100)/100
#         long_val = int(long_val*100)/100
        
        print('위도:{}, \t경도:{}'.format(lati_val, long_val))
        folium.Marker([lati_val,long_val], popup=str(idx)
                     ).add_to(map_3)    
        
        idx += 1
        
    if(idx > 7):
        break;
        
map_3
```
